```python
new_df = df.drop(columns=['date', 'lights'])
```

```python
scaler = MinMaxScaler()
normalised_df = pd.DataFrame(scaler.fit_transform(new_df), columns=new_df.columns)
```

```python
target= normalised_df['Appliances']
normalised_df.drop(columns=['Appliances'], inplace=True)
```

```python
##split the data
X1_train, X1_test, y1_train, y1_test = train_test_split(normalised_df, target, test_size=0.3, random_state=42)
```

```python
model1 = LinearRegression().fit(X1_train, y1_train)
```

```python
predicted_values = model1.predict(X1_test)
```

```python
mae = mean_absolute_error(y1_test, predicted_values)
```

```python
print(" mean absolute error is %.2f" %round(mae,2))
```

```python
rss = np.sum(np.square(y1_test - predicted_values))
```

```python
print(" Residual Sum of Squares is %.2f" %round(rss,2))
```

```python
rmse = np.sqrt(mean_squared_error(y1_test, predicted_values))
```

```python
print(" Residual Sum of Squares is %.2f" %round(rmse,3))
```

```python
r_sq = model1.score(X1_train, y1_train)
```

```python
print(" The Coefficient of Determination is %.2f" %round(r_sq,2))
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(X1_train, y1_train)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(X1_train, y1_train)

#comparing the effects of regularisation
def get_weights_df(model, feat, col_name):
    #this function returns the weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(model1, X1_train, 'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, X1_train, 'Ridge_Weight')
lasso_weights_df = get_weights_df(lasso_reg, X1_train, 'Lasso_weight')
final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
```
